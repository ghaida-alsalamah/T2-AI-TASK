## Feature Engineering

```python
df["has_description"] = df["description"].notna().astype(int)
df["has_hashtags"] = df["hashtags"].notna().astype(int)

df = df.drop(columns=["description", "hashtags"])
```

```python
df.isnull().sum()
```

```python
df["like_rate"] = df["likes"] / (df["views"] + 1)
df["comment_rate"] = df["comments"] / (df["views"] + 1)
```

```python
df["is_hd"] = (df["height"] >= 720).astype(int)
```

```python
df["short_vid"] = (df["duration"] < 1200).astype(int)        # < 20 mim
df["long_vid"] = (df["duration"] >= 1200).astype(int)        # > 20 min
```

```python
df = df.drop(columns=["video_id", "url", "title"])
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=["codec"], drop_first=True)
```

```python
df = pd.get_dummies(df, columns=["category"], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

numuric_featuers = ["duration", "bitrate", "height", "width", "frame rate", "frame rate(est.)"]

scaler = StandardScaler()
df[numuric_featuers] = scaler.fit_transform(df[numuric_featuers])

df[numuric_featuers].head()
```

```python
df["popularity_score"] = (0.5*np.log1p(df["views"]) + 0.3*np.log1p(df["likes"]) + 0.2*np.log1p(df["comments"]))
```

```python
df.columns
```

## Data Augmentation

```python
continuous_features = ["duration", "bitrate", "bitrate(video)", "height", "width", "frame rate",
                       "frame rate(est.)", "like_rate", "comment_rate", "popularity_score" ]
```

```python
augmented_df = df.sample(n=5000, replace=True, random_state=42).copy()
```

```python
for col in continuous_features:
    noise = np.random.normal( loc=0, scale=0.01 * augmented_df[col].std(), size=augmented_df.shape[0] )
    augmented_df[col] = augmented_df[col] + noise
```

```python
augmented_df[continuous_features] = augmented_df[continuous_features].clip(lower=0)
```

```python
augmented_df.shape
```

```python
augmented_df.head()
```

```python
augmented_df.tail()
```

```python
df_augmented = pd.concat([df, augmented_df], ignore_index=True)
```

```python
df.shape, df_augmented.shape
```
